```python
import pandas as pd
import numpy as np
import json


data = pd.read_csv('../data/wordle_remake.csv')

word_length = 4
for col in data.columns:
    ## Convert to Lowercase
    data[col] = data[col].str.lower()

    ## Remove words that do not match the expected length
    data.loc[data[col].str.len() != word_length, col] = None
    word_length += 1

print(data)

output = {}
for col in data.columns:
    words = data[col].dropna().tolist()
    output[col] = words

data.to_csv('../data/wordle_remake.csv', index=False)

with open('../data/wordle_remake.json', 'w') as f:
    json.dump(output, f, indent=2)
```
